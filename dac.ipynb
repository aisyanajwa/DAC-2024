```python
#combine solar 2014, 2015, 2016, 2017 using concat
solar_all = pd.concat([solar_2014, solar_2015, solar_2016, solar_2017])
solar_all
```

```python
# check maximum of Minute column
solar_all['Minute'].max()
```

```python
# Format the datetime objects
solar_all['Datetime'] = pd.to_datetime(solar_all[['Year', 'Month', 'Day', 'Hour']])
solar_all
```

```python
# Change format date_time in weather dataset then rename columns date_time to Datetime. It will be used as primary key
weather['date_time'] = pd.to_datetime(weather['date_time'])
weather = weather.rename(columns={'date_time': 'Datetime'})
weather['Datetime'] #retrieve data but  only for Datetime
```

```python
# combine solar_all with weather on primary key Datetime using merge left
solar = pd.merge(solar_all, weather, on='Datetime', how='left')
solar
```

```python
# Move the Datetime column to the far left
cols = ['Datetime'] + [col for col in solar.columns if col != 'Datetime']

# Rename Datetime column in solar to Timestamp to match the expected output
solar = solar[cols]
solar = solar.rename(columns={'Datetime':'Timestamp'})
solar
```

```python
# Change format Timestamp type to match the expected output
train['Timestamp'] = pd.to_datetime(train['Timestamp'], format='%b %d, %Y %I%p')

train['Timestamp'] #retrieve data only for Timestamp
```

```python
# Combine train and solar on primary key Timestamp using left join
final_data = pd.merge(train, solar, on='Timestamp', how='left')

# Display the merged DataFrame
final_data
```

```python
# check maximum of hour to understanding data
final_data['Hour'].max()
```

```python
# Check for missing values and sum them to identify which columns are missing data and need to be handled
final_data.isnull().sum()
```

```python
# check all of name columns
final_data.columns
```

```python
# Filling NaN values in final_data using the KNN method

from sklearn.impute import KNNImputer #import packages

# Assume final data is original dataframe
# Separate numeric and non numeric columns
df_numeric = final_data.select_dtypes(include='number')
df_non_numeric = final_data.select_dtypes(exclude='number')

# Imputation using KNN
imputer = KNNImputer(n_neighbors=5) #parameter specifies that the imputer will use the 5 nearest neighbors to estimate missing values (k=5)

# Applies KNN inputation to the numeric column in final data. Learns(fit) and transform(applies) model, ensure that columns and index are preserved
df_numeric_imputed = pd.DataFrame(imputer.fit_transform(df_numeric), columns=df_numeric.columns, index=df_numeric.index)

# Combine the non-numeric columns with the numeric columns that have been imputed using KNN along the columns (axis=1)
final_data_KNN = pd.concat([df_non_numeric, df_numeric_imputed], axis=1)

# Retrieve data
final_data_KNN
```

```python
# Analysis error with calculate DHI - DHI and use absolute
error = (final_data['DHI'] - final_data_KNN['DHI']).abs()
print('Rata-rata Kesalahan Absolut:', error.mean())
```

```python
# Fill NaN with No Data and retrieve
final_data_KNN['Cloud Type'].fillna('No Data', inplace=True)
final_data_KNN
```

```python
# change Format Timestamp in Test and retrieve
test['Timestamp'] = pd.to_datetime(test['Timestamp'], format='%b %d, %Y %I%p')

test['Timestamp']
```

```python
# Filling NaN values in solar using the KNN method

from sklearn.impute import KNNImputer #imput packages

# Assume final data is original dataframe
# Separate numeric and non numeric columns
df_numeric = solar.select_dtypes(include='number')
df_non_numeric = solar.select_dtypes(exclude='number')

# Imputation using KNN
imputer = KNNImputer(n_neighbors=5) #parameter specifies that the imputer will use the 5 nearest neighbors to estimate missing values (k=5)

# Applies KNN inputation to the numeric column in final data. Learns(fit) and transform(applies) model, ensure that columns and index are preserved
df_numeric_imputed = pd.DataFrame(imputer.fit_transform(df_numeric), columns=df_numeric.columns, index=df_numeric.index)

# Combine the non-numeric columns with the numeric columns that have been imputed using KNN along the columns (axis=1)
solar_KNN = pd.concat([df_non_numeric, df_numeric_imputed], axis=1)

# Retrieve data
solar_KNN
```

```python
# Fill NaN with No Data and retrieve
solar_KNN['Cloud Type'].fillna('No Data', inplace=True)
solar_KNN
```

```python
# Combine test and solar_KNN on primary key Timestamp using merge and retrieve data
final_test = pd.merge(test, solar_KNN, on='Timestamp')
final_test
```

```python
# Check missing values and sum
final_test.isnull().sum()
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor #to use Random Forest Regression
from sklearn.model_selection import train_test_split #to separate between train and test
from sklearn.metrics import mean_squared_error #to evaluation model performance
import numpy as np

# Assume final_data_KNN is the imputed training data
# Separates fitur and target for the training data
# solar_KNN -> test, final_data_KNN -> train

X_train_data = final_data_KNN.drop(['% Baseline', 'Timestamp'], axis=1)
# Drop Baseline and Timestamp column and drop column(axis = 1)
# Output from x contains only the independent variables (feature columns) that will train the model

y_train_data = final_data_KNN['% Baseline']
# Output from y contains the dependent variable (target column) that the model is trying to predict.

# Identify non numeric columns and encoding with One-Hot Encoding

# Select non numeric column and save it in non_numeric_cols
non_numeric_cols = X_train_data.select_dtypes(exclude=['number']).columns

# Create dummy columns for each category in non-numeric columns
# Drop first dummy columns to avoid multicollinearity(a situation where several dummy columns can be highly correlated with each other)
X_train_encoded = pd.get_dummies(X_train_data, columns=non_numeric_cols, drop_first=True)

# Separate data to train and test, use test size 20% and random_state 42
X_train, X_test, y_train, y_test = train_test_split(X_train_encoded, y_train_data, test_size=0.2, random_state=42)

# Create Random Forest model
# Decide estimators 100(general) and random_state in 42(general)
model_rf = RandomForestRegressor(n_estimators=100, random_state=42)

# train model, x_train = for training the model, y_train = data that will be trained and predicted
model_rf.fit(X_train, y_train)

# Make predictions on the test data
# Output from y_pred_rf is an array containing the predicted values ​​from the model for each instance in X_test.
y_pred_rf = model_rf.predict(X_test)

# Evaluate model using RMSE and retreive data
rmse_rf = np.sqrt(mean_squared_error(y_test, y_pred_rf))
print('Root Mean Squared Error dengan Random Forest:', rmse_rf)
```

```python
# Preprocess final_test in the same way and ignore errors
X_final_test = final_test.drop(['% Baseline', 'Timestamp'], axis=1, errors='ignore')

# Use One-hot encoding
# Change column non_numeric with dummy columns and drop first dummy column to avoid multicollinearity
X_final_test_encoded = pd.get_dummies(X_final_test, columns=non_numeric_cols, drop_first=True)

# Ensure features in final_test has same column with train dataset
# Add missing columns with zeros
# Calculate the difference between the columns in the training data and test data
missing_cols = set(X_train_encoded.columns) - set(X_final_test_encoded.columns)
# missing_cols is a column that is available in `X_train` but not in `X_final_test`
#Therefore, the column will be missing and should be added back to ensure consistent data

# Create new column name col in X_final_test and fill it with 0
for col in missing_cols:
    X_final_test_encoded[col] = 0

# Sort column X_final_test same as X_train
X_final_test_encoded = X_final_test_encoded[X_train_encoded.columns]

# Create prediction, output y_test_pred_rf is an array containing predicted values ​​for test data
y_test_pred_rf = model_rf.predict(X_final_test_encoded)

# Add the prediction results to the final_test DataFrame, so the new column shows the model predictions for each instance in the test data
final_test['Predicted % Baseline'] = y_test_pred_rf

# Display results
print(final_test.head(5))
```

```python
# Calculating the importance of features from a trained model
importances = model_rf.feature_importances_ # Output importances is an array containing the importance value of each feature used by the model.

# saves the feature names of the encoded training data (X_train_encoded)
feature_names = X_train_encoded.columns # Output feature_names is a list of feature column names that correspond to values ​​in importances.

# Create DataFrames for important features
feature_importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': importances
})

# Sort DataFrame based importance value and based on descending and take top 8 features using head
top_8_features = feature_importance_df.sort_values(by='Importance', ascending=False).head(8)

# Display results
print(top_8_features)
```
